## Un poco de historia

Hace 40 a√±os, un grupo de unos 50 pensadores se tomaron a si mismos una foto en el c√©sped de la casa de Endicott del MIT. Unos pocos en la conferencia de f√≠sica de la computaci√≥n, organizada entre el MIT e IBM, pensaron que estaban haciendo historia en 1981. Podr√≠a decirse que fue el lugar de nacimiento de la f√≠sica de la computaci√≥n, y especialmente donde floreci√≥ el campo de la computaci√≥n cu√°ntica como un tema serio, digno de ser tratado en libros universitarios.

Fue en esta conferencia donde Feynman pronunci√≥ su famosa cita

>  ‚ÄúNature isn‚Äôt classical, dammit, and if you want to make a simulation of nature, you‚Äôd better make it quantum mechanical, and by golly it‚Äôs a wonderful problem, because it doesn‚Äôt look so easy.‚Äù - Richard Feynman


![](https://github.com/qiskit-community/ibm-quantum-challenge-2021/raw/7794356e7ae7fb7569c853747c6791e443cb6fc9/content/ex1/resources/conference-photo.jpeg)

Uno de los temas que se discutieron en la conferencia fue la **computaci√≥n reversible** o **reversibilidad de los c√≥mputos**, en la cual Tommaso Toffoli y Edward Fredkin del MIT estuvieron pensando sobre ella ya a√±os antes. Toffoli ide√≥ una versi√≥n reversible de las compuertas **AND/NAND**, la cual ahora recibe el nombre de **Compuerta de Toffoli** o **Controlled-Controlled-NOT**. Como sabemos, la compuerta **NAND** es universal en computaci√≥n cl√°sica (a partir de ella podemos construir el resto de compuertas), y por lo tanto, la compuerta de Toffoli es una compuerta l√≥gica **universal y reversible**.

De hecho, la computaci√≥n cu√°ntica es una forma especial de computaci√≥n reversible; cualqueir compuerta reversible puede ser implementada en una computadora cu√°ntica y por lo tanto, la compuerta de Toffoli es tambi√©n una compuerta cu√°ntica. Sin embargo, la compuerta Toffoli **NO es una compuerta universal para la computaci√≥n cu√°ntica**.

## Compuertas l√≥gicas cl√°sicas

Uno de los modelos m√°s utilizados en computaci√≥n cl√°sica es la L√≥gica Booleana o la compuertas l√≥gicas cl√°sicas. Tales compuertas representa funciones booleanas, funciones que solo tienen inputs y outputs binarios (0, 1). Un aspecto interesante de la l√≥gica booleana es que, todas las posibles funciones posibles pueden ser formadas √∫nicamente utilizando una peque√±a combinaci√≥n de compuertas l√≥gicas. De tal forma, que a ese reducido set de compuertas les llamamos **conjuntos funcionalmente completos**. Uno de los m√°s famosos es el conformado por la **AND** y la **NOT**, con estas dos compuertas podemos expresar todas las funciones posibles. Tambi√©n se cumple para el caso **OR y NOT**. Existen conjuntos m√°s peque√±os, como el **NAND** y **OR** los cuales por s√≠ solos ya son universales, sin embargo las funciones **AND**,**NOT** Y **OR** suelen ser vistas como los bloques o pilares b√°sicos de la computaci√≥n cl√°sica.

## ¬øQu√© son los circuitos cu√°nticos?

Los circuitos cu√°nticos son modelos de computaci√≥n cu√°ntica los cuales nuestra computaci√≥n realizada es una secuencia de compuertas cu√°nticas. Las compuertas cu√°nticas suelen ser representadas como rotaciones en la esfera de Bloch. Echemos un vistazo a algunas delas compuertas m√°s populares:

### Compuerta X

Esta compuerta est√° representada por la Matriz de Pauli X:

$$X = \begin{pmatrix}
0 & 1 \\
1 & 0 \\
\end{pmatrix}$$

Y lo qu√© hace esta compuerta es efectuar rotaciones de $\pi$ rad alrededor del eje $X$ en nuestra esfera de Bloch. Es decir, cambia nuestros estados de $|0\rangle$ a $|1\rangle$ y de $|1\rangle$ a $|0\rangle$. Es el equivalente cu√°ntico de la compuerta cl√°sica **NOT** y a veces es llamada como **bit-flip**.

```python
x_gate = QuantumCircuit(1) # Creamos un circuito de 1 qubit
x_gate.x(0) # Aplicamos una compuerta X en el bit con √≠ndice 0 (el primero y √∫nico)
x_gate.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator') # Seleccionamos el simulador
# Ejecutamos el circuito en el simulador
# y obtenemos el resultado y de ah√≠ el vector de estado
result = execute(x_gate , backend).result().get_statevector() 
# Dibujamos en una esfera de Bloch
plot_bloch_multivector(result)
```
